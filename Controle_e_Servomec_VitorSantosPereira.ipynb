```python
import matplotlib.pyplot
```

```python
class Besouro:

    def __init__(self):

        self.L_t = 1
        self.P_t = 0
        self.A_t = 0

        self.L_t1 = 0
        self.P_t1 = 0
        self.A_t1 = 0
```

```python
#Instancia Besouro

inseto = Besouro()
```

```python
#valores iniciais:

b = 1.2 #taxa de natalidade
ul = 0.4 #taxa de mortalidade das larvas
up = 0.5 #taxa de mortalidade das pupas
ua = 0.2 #taxa de mortalidade dos adultos
t = 280 #tempo em dias

dias = []
populacao = []
```

```python
#difinições das funções:

def biologicSystem():
  inseto.L_t1 = b * inseto.A_t
  inseto.P_t1 = inseto.L_t * (1 - ul)
  inseto.A_t1 = inseto.P_t * (1 - up) + inseto.A_t * (1 - ua)

def update_biologicSystem():
  inseto.L_t = inseto.L_t1
  inseto.P_t = inseto.P_t1
  inseto.A_t = inseto.A_t1

def print_inseto():
  print("larvas: ", inseto.L_t)
  print("pupa: ", inseto.P_t)
  print("adulto: ", inseto.A_t)
  print("\n")
```

```python
b = 1.5

for i in range(1, t+1):
  biologicSystem()
  update_biologicSystem()
  populacao.append(inseto.A_t)
  dias.append(i)
```
