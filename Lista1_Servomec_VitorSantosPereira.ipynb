#Questão 1


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class Besouro:

    def __init__(self):

        self.L_t = 1
        self.P_t = 0
        self.A_t = 0

        self.L_t1 = 0
        self.P_t1 = 0
        self.A_t1 = 0

b = 0 #taxa de natalidade
ul = 0 #taxa de mortalidade das larvas
up = 0 #taxa de mortalidade das pupas
ua = 0 #taxa de mortalidade dos adultos
t = 40 #tempo em semanas
```

```python
#Instancia Besouro

inseto = Besouro()
```

```python
#difinições das funções:

def biologicSystem():
  inseto.L_t1 = b * inseto.A_t
  inseto.P_t1 = inseto.L_t * (1 - ul)
  inseto.A_t1 = inseto.P_t * (1 - up) + inseto.A_t * (1 - ua)

def update_biologicSystem():
  inseto.L_t = inseto.L_t1
  inseto.P_t = inseto.P_t1
  inseto.A_t = inseto.A_t1

def print_inseto():
  print("larvas: ", inseto.L_t)
  print("pupa: ", inseto.P_t)
  print("adulto: ", inseto.A_t)
  print("\n")
  print("larvas t+1: ", inseto.L_t1)
  print("pupa t+1: ", inseto.P_t1)
  print("adulto t+1: ", inseto.A_t1)
  print("\n")

def reset_inseto():
  inseto.L_t = 1
  inseto.P_t = 0
  inseto.A_t = 0
  inseto.L_t1 = 0
  inseto.P_t1 = 0
  inseto.A_t1 = 0
```

## Fim da População

```python
reset_inseto()
```

```python
#valores iniciais:

b = 1 #taxa de natalidade
ul = 0.3 #taxa de mortalidade das larvas
up = 0.5 #taxa de mortalidade das pupas
ua = 0.6 #taxa de mortalidade dos adultos

dias = []
populacao_larvas = []
populacao_pupas = []
populacao_adultos = []
```

```python
for i in range(1, t+1):
  biologicSystem()
  update_biologicSystem()

  populacao_larvas.append(inseto.L_t)
  populacao_pupas.append(inseto.P_t)
  populacao_adultos.append(inseto.A_t)

  dias.append(i)
```

```python
df = pd.DataFrame({
    "Larvas": populacao_larvas,
    "Pupas" : populacao_pupas,
    "Adultos": populacao_adultos
})
plt.plot(df)
plt.title("Mort. Larvas: " + str(ul) + "\n" + "Mort. Pupas: " + str(up) + "\n" + "Mort. Adultos: " + str(ua) + "\n" + "Natalidade: " + str(b))
plt.xlabel("Tempo em Semanas")
plt.ylabel("Unidades de Espécie")
legend = plt.legend(['Larvas','Pupas','Adultos'], title = "Legendas")
```

Com a alta taxa de mortalidade de larvas, pupas e adultos, a taxa de natalidade acaba não suportando e a população entra em extinção.

```python
#difinições das funções:

def biologicSystem():
  inseto.L_t1 = b * inseto.A_t
  inseto.P_t1 = inseto.L_t * (1 - ul - (uag * Ar))
  inseto.A_t1 = inseto.P_t * (1 - up) + inseto.A_t * (1 - ua)

def update_biologicSystem():
  inseto.L_t = inseto.L_t1
  inseto.P_t = inseto.P_t1
  inseto.A_t = inseto.A_t1

def print_inseto():
  print("larvas: ", inseto.L_t)
  print("pupa: ", inseto.P_t)
  print("adulto: ", inseto.A_t)
  print("\n")
  print("larvas t+1: ", inseto.L_t1)
  print("pupa t+1: ", inseto.P_t1)
  print("adulto t+1: ", inseto.A_t1)
  print("\n")

def reset_inseto():
  inseto.L_t = 1
  inseto.P_t = 0
  inseto.A_t = 0
  inseto.L_t1 = 0
  inseto.P_t1 = 0
  inseto.A_t1 = 0
```

#Questão 3

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.integrate import odeint
```

```python
#Valores Iniciais

a = 0.311
b = 0.518
c = 1.036
d = 0.311
e = 0.161
f = 4.599
g = 2.469
h = 0.322

t = np.linspace(0, 10, 101)
```

```python
#Funções

def model(diff_array, t):
    x,y,z= diff_array
    dxdt = x*(1-x) - ((x*y)/(x+a))
    dydt = -b*y + ((c*x*y)/(x+d)) - ((y*z)/(y+e))
    dzdt = (f*z**2) - ((g*z**2)/(y+h))

    return [dxdt, dydt, dzdt]

def renderGraph(x, y, z):    
    result = odeint(model, [x, y, z], t) #Resolve as equações diferenciais

    df = pd.DataFrame({
      "Presa": result[:, 0],
      "Predador Especialista" : result[:, 1],
      "Predador Generalista": result[:, 2]
    })

    plt.plot(df)
    plt.xlabel("Tempo em Estantes")
    plt.ylabel("Unidades de Espécie")
    legend = plt.legend(['Presa','Predador Especialista','Predador Generalista'], title = "Legendas")

    plt.title(
        'Modelo de Cadeia Alimentar\n'
        + ' x = ' + str(x) + ' y = ' + str(y) + ' z = ' + str(z) 
        )
```

```python
renderGraph(0.2, 0.1, 0.7)
```

```python
renderGraph(0.7, 0.5, 0.8)
```

```python
renderGraph(0.8, 0.1, 0.1)
```
